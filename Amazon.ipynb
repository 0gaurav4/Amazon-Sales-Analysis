## <center>Data Visualization</center>

### Gender

```python
a = sns.histplot(x='Gender', data=df, hue='Gender')

# Adding title
a.set_title('Distribution of Gender')
# Adding count labels on top of the bars
for bars in a.containers:
    a.bar_label(bars)

# Display the plot
plt.show()
```

### Gender X Amount Expended

```python
b= df.groupby(['Gender'], as_index=False)['Amount'].sum()
a = sns.barplot(x='Gender', y='Amount', data= b, hue='Gender')

# Adding title
a.set_title('Amount Expended')
# Adding count labels on top of the bars
for bars in a.containers:
    a.bar_label(bars)

# Display the plot
plt.show()
```

### Age Group

```python
s = sns.countplot(x='Age Group', data=df,hue= 'Gender')
s.set_title('Distribution of Age Group')
for bars in s.containers:
    s.bar_label(bars)
plt.show() 
```

### Age Group X Amount Expended

```python
c= df.groupby(['Age Group'], as_index=False)['Amount'].sum()
s = sns.barplot(x='Age Group',y='Amount', data=c)
s.set_title('Age Group X Amount')
sns.set(rc={'figure.figsize':(10,3)})
for bars in s.containers:
    s.bar_label(bars)
plt.show()
```

### State X Orders

```python
d= df.groupby(['State'], as_index=False)['Orders'].sum()
u= sns.barplot(x='State' , y= 'Orders', data=d)
sns.set(rc={'figure.figsize':(18,5)})
for bars in u.containers:
    u.bar_label(bars)
    
plt.show()
```

### Marital Status

```python
m = sns.countplot(data = df, x = 'Marital_Status')
m.set_title('Married= 0, Unmarried= 1')
sns.set(rc={'figure.figsize':(10,9)})
for bars in m.containers:
    m.bar_label(bars)
```

### Gender X Marital Status X Amount Expended

```python
e= df.groupby(['Gender','Marital_Status'], as_index=False)['Amount'].sum()
o= sns.barplot(x='Marital_Status' , y='Amount', hue= 'Gender',data=e)
sns.set(rc={'figure.figsize':(7,5)})
o.set_title('Married= 0, Unmarried= 1')
for bars in o.containers:
    o.bar_label(bars)
```

## Occupation

```python
j= sns.countplot(x='Occupation', data=df )
sns.set(rc={'figure.figsize':(20,5)})
for bars in j.containers:
    j.bar_label(bars)
    plt.show()
```

### Occupation X Amount Expended

```python
f= df.groupby(['Occupation'], as_index=False)['Amount'].sum()
k= sns.barplot(x='Occupation',y='Amount',data=f)
for bars in k.containers:
    k.bar_label(bars)
    plt.show()
```

### Product Category

```python
g= sns.countplot(x='Product_Category', data=df)
sns.set(rc={'figure.figsize':(30,5)})
for bars in g.containers:
    g.bar_label(bars)
    plt.show()
```

### Product Categories X Orders

```python
m= df.groupby(['Product_Category'], as_index=False)['Orders'].sum()
sns.set(rc={'figure.figsize':(30,7)})
h= sns.barplot(x='Product_Category' ,y='Orders',data=m)
for bars in h.containers:
    h.bar_label(bars)
    plt.show()
```

## <center>Conclusions:</center>

- Most of the buyers are working in IT, Healthcare and Aviation sector. 
- Most of the buyers are from age group 26-35 yrs.
- Most of the costumers married women from UP, Maharastra and Karnataka.
- Most of the sold products are from Clothing,food,Electronics and footwear category.
